```python
import seaborn as sb
import pandas as pd
import re
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("meteo-nogal-09.csv",engine = "python") 
```

```python
#Me fijo como se llaman las columnas que no se ocupan
df.columns
```

```python
#Obtengo el nombre exacto de las columnas que no se ocupan para despues eliminarlas
Unnames = []
for i in range(0,len(df.columns)):
    if(re.search("^Unnamed",df.columns[i]) != None):
        Unnames.append(df.columns[i])
Unnames
```

```python
df.head()
```

```python
#Eliminar las columnas Unnamed y el primer renglon ya que Ã©ste no aporta datos. AdemÃ¡s reviso el tipo los datos de cada columna
df = df.drop(Unnames,1)
df = df.drop(0,0)
df.dtypes
```

```python
#Recorto los datos para solamente utilizar los datos de 2009, al igual que convierto la columna DATE a formato fecha
df["DATE"] = pd.to_datetime(df["DATE"])
df = df[df["DATE"] < "2010-01-01"]
df.tail()
```

```python
#Convertir las variables de object a float64
tempD = df["DATE"]
tempT = df["TIME"]
df[df.columns] = df[df.columns].apply(pd.to_numeric, errors='coerce')
df["DATE"] = tempD
df["TIME"] = tempT
df.dtypes
```

```python
#Me aseguro de que todo estÃ¡ bien
df.head()
```

```python
#Construyo la matriz de correlacion
corr = df.corr()
```

```python
corr
```

```python
sb.heatmap(corr,cmap="viridis",robust=True,square=True,linewidths=.01)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_xticks(np.arange(len(corr)))
ax.set_yticks(np.arange(len(corr)))

ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)

plt.setp(ax.get_xticklabels(), rotation=90,ha="right",rotation_mode="anchor")


plt.imshow(corr, cmap='viridis',interpolation='nearest')
plt.colorbar()

plt.show()
```

```python
def array(arr,st,fin):
    arra = []
    for i in range(st,fin):
        arra.append(arr[i])
    return arra
```

```python
Cols = []
n = -1
for i in corr.index:
    n=n+1
    for j in array(corr.index,n,len(corr)):
        if(abs(corr[i][j]) > 0.6 and i != j):
            Cols.append([i,j])
```

```python
for i in range(0,len(Cols)):
    plt.figure(figsize=(15,5))
    sb_plot = sb.regplot(x = df[Cols[i][0]], y = df[Cols[i][1]],fit_reg=False)
    fig = sb_plot.get_figure()
    fig.savefig(fname = Cols[i][0] + " vs " + Cols[i][1] ,plt=2000)
    plt.grid()
    plt.show()
```
